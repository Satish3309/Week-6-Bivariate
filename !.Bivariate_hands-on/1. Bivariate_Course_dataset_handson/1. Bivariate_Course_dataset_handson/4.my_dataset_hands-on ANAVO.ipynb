```python
import pandas as pd

# Create the reshaped dataset directly
data_long = {
    "Name of chef": ["Saravanan", "Mahendran", "Sima",
                     "Saravanan", "Mahendran", "Sima",
                     "Saravanan", "Mahendran", "Sima"],
    "Dish": ["Biriyani", "Biriyani", "Biriyani",
             "Grill", "Grill", "Grill",
             "Mutton", "Mutton", "Mutton"],
    "Gender":["Male","Male","Female",
              "Male","Male","Female",
              "Male","Male","Female"],
    "Score": [88, 80, 92,
              82, 86, 78,
              85, 83, 90]}

# Convert to DataFrame
df_long = pd.DataFrame(data_long)

print(df_long)
```

```python
#Now we have dependent variable so lets find out how chef gender and dish affect the score.
```

```python
import statsmodels.api as sm #(import libraries.)
from statsmodels.formula.api import ols

model = ols('Score ~ C(Gender) + C(Dish) + C(Gender):C(Dish)', data=df_long).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

#how to write concolusion
#pvalue of gender is > 0.05 so we accept Ho (Accept null hypothesis) there is no significant difference between gender and score
#pvalue of Dish is > 0.05 so we accept Ho (Accept null hypothesis) there is no significant difference between Dish and score
#pvalue of gender:Dish is > 0.05 so we accept Ho (Accept null hypothesis) there is no significant 
#difference between gender:Dish  and score
