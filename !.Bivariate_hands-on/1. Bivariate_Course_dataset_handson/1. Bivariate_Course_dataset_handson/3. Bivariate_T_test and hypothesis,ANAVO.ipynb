#hypothesis
question: is ssc_p of male and hsc_p of male is same?
#Thumb rule is if p < or = 0.05 reject null hypothesis and Accept Alter hypothesis. 

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna() #(null values not work for T-Test)
Male=dataset[dataset["gender"]=="M"]["ssc_p"]
Male1=dataset[dataset["gender"]=="M"]["hsc_p"]

ttest_rel(Male,Male1)
```

```python
#in above problem we have p value is 0.49 so condition is falls (0.49 not < 0.05)
#so we reject Alter hypothesis and accept null hypothesis. 
#Answer: ssc_p of male and hsc_p of male is Not same.(Null hypothesis)
```

```python
import statsmodels.api as sm #(import libraries.)
from statsmodels.formula.api import ols

model = ols('salary ~ C(gender) + C(specialisation) + C(gender):C(specialisation)', data=dataset).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
Interpretation.
#pvalue of gender is 0.08 > 0.05 so we accept Ho (Accept null hypothesis) 
#there is no significant difference between gender and salary

#pvalue of specialisation is 0.000081 < 0.05 so we accept H1 (Reject null hypothesis)
#there is a significant difference between specialisation and salary.

#pvalue of (gender):(specialisation) is 0.917733 > 0.05 so we accept Ho (Accept null hypothesis) there is no significant 
#difference between (gender):(specialisation  and salary.
```

In other words.

Gender (p = 0.08 > 0.05)

Fail to reject the null hypothesis.

Conclusion: Salary does not significantly differ between male and female candidates.

Specialisation (p = 0.000081 < 0.05)

Reject the null hypothesis.

Conclusion: Salary does significantly differ based on MBA specialisation (e.g., Marketing & HR vs Marketing & Finance).

Interaction (Gender × Specialisation, p = 0.917733 > 0.05)

Fail to reject the null hypothesis.

Conclusion: The effect of specialisation on salary does not depend on gender. In other words, specialisation influences salary equally for both males and females.
