```python
dataset = pd.read_csv("Placement.csv")

# Force conversion of numeric-like columns
numeric_cols = ["ssc_p", "hsc_p", "degree_p", "etest_p", "mba_p", "salary"]
for col in numeric_cols:
    dataset[col] = pd.to_numeric(dataset[col], errors="coerce")
```

```python
numeric_cols = ["ssc_p", "hsc_p", "degree_p", "etest_p", "mba_p", "salary"]

for col in numeric_cols:
    Re_Dataset[col] = pd.to_numeric(Re_Dataset[col], errors="coerce")
```

```python
from sklearn.impute import SimpleImputer
```

```python
salary_imputer = SimpleImputer(strategy="constant", fill_value=0)
Re_Dataset["salary"] = salary_imputer.fit_transform(Re_Dataset[["salary"]])
```

```python
median_imputer = SimpleImputer(strategy="median")
Re_Dataset[quan] = median_imputer.fit_transform(Re_Dataset[quan])
```

```python
mean_imputer = SimpleImputer(strategy="most_frequent")
Re_Dataset[qual] = mean_imputer.fit_transform(Re_Dataset[qual])
```
